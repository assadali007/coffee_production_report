### Data Prep

```python
axes_settings= {
    "spines.top": False,
    "spines.right": False,
}
plt.rc("axes",**axes_settings)
```

```python
brazil_vs_others = coffee_production.assign(
    world_total = coffee_production.drop('Brazil',axis=1).sum(axis=1)
).loc[:,['Brazil','world_total']]
brazil_vs_others.head(5)
```

```python
fig,ax = plt.subplots(figsize=(12,2))
ax.stackplot(
    brazil_vs_others.index,
    brazil_vs_others['Brazil'],
    brazil_vs_others['world_total'],
    colors= ['#009739','grey']
);
fig.suptitle('Brazil share of Market Grows as Global output surges')
ax.set_ylabel('Production k 60kgs Bags')
ax.set_xticks(brazil_vs_others.index[2::4]);
ax.legend(['Brazil','world total'])
plt.show()
```

```python
pie_series =brazil_vs_others.iloc[0].sort_values(ascending=False).to_frame()
pie_series
```

```python
fig, ax = plt.subplots()

ax.pie(
    pie_series['1990'],
    startangle=90,
    labels = ["",""],
    #autopct='%.0f%%',
    pctdistance=.85,
    colors=["White", "#009739"]
);

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# # Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round((brazil_vs_others.loc['1990', 'Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 1990", fontsize=14)

plt.show()
```

```python
pie_series_2 =brazil_vs_others.iloc[-1].sort_values(ascending=False).to_frame()
pie_series_2
```

```python
fig, ax = plt.subplots()

ax.pie(
    pie_series_2['2018'],
    startangle=90,
    labels = ["",""],
    #autopct='%.0f%%',
    pctdistance=.85,
    colors=["White", "#009739"]
);

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# # Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 2018", fontsize=14)

plt.show()
```

# Subplots

* Read in `prices-paid-to-growers.csv` and create a subplot grid of histograms showing the distribution of prices paid to the countries Brazil, Colombia, Ethiopia, and one for all other nations.

```python
prices_paid_to_growers = pd.read_csv("../Data/coffee/prices-paid-to-growers.csv").T.drop(9, axis=1)

prices_paid_to_growers
prices_paid_to_growers.columns= prices_paid_to_growers.iloc[0]

prices_paid_to_growers.drop("prices_paid_to_growers", inplace=True)

prices_paid_to_growers["Other Nations"] =  prices_paid_to_growers.drop(["Colombia", "Brazil", "Ethiopia"], axis=1).mean(axis=1)
```

```python
prices_paid_to_growers.loc[:,['Brazil','Colombia','Ethiopia','Other Nations']].head(5)
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,10),sharex='all',sharey='all')
ax[0][0].hist(prices_paid_to_growers['Brazil']);
ax[0][0].set_title('Brazil');

ax[0][1].hist(prices_paid_to_growers['Colombia']);
ax[0][1].set_title('Colombia');

ax[1][0].hist(prices_paid_to_growers['Ethiopia']);
ax[1][0].set_title('Ethiopia');

ax[1][1].hist(prices_paid_to_growers['Other Nations']);
ax[1][1].set_title('Other Nations');
```
